```python
import pandas as pd

df = pd.read_csv('/Users/pedrovallocci/Documents/PhD (local)/Research/By Topic/Bargaining Power/data/msa_emp_tickers_manual_changes.csv')
dog = df[['unemp_total', 'msa']]
msa_dict = dict(dog.values)
```

```python
msa_ticker_list = list(df['unemp_total'])
msa_name_list = list(df['msa'])
```

```python
import requests
import json
headers = {'Content-type': 'application/json'}
maxlen = len(msa_ticker_list)
i = 0
stop = False
json_data = []
while True:
    ini = 50*i
    print(ini)
    end = min(ini+50, maxlen)
    sublist = msa_ticker_list[ini:end]
    data = json.dumps({"seriesid": sublist,"catalog": True, "startyear":"2010", "endyear":"2018", "calculations": True,  "annualaverage": True, "registrationkey":"b0d116e1115f44329937c5c105cc44a7"})
    p = requests.post('https://api.bls.gov/publicAPI/v2/timeseries/data/', data=data, headers=headers)
    json_data.append(json.loads(p.text))
    i = i+1
    if ini+50 > maxlen:
        break
```

```python
import numpy as np
msa_vector = []
seriesid_vector = []
year_vector = []
period_vector = []
value_vector = []
i = 0
while True: 
    my_json_data = json_data[i]
    for series in my_json_data['Results']['series']:
    #     x=prettytable.PrettyTable(["series id","year","period","value","footnotes"])
    #     series_name = df.loc[df['unemp_rate'] == seriesId]['msa'][1]
        for item in series['data']:
            msa_vector.append(msa_dict[series['seriesID']])
            seriesid_vector.append(series['seriesID'])
            year_vector.append(item['year'])
            period_vector.append(item['period'])
            value_vector.append(item['value'])
    i = i+1
    print(i)
    if i >= np.ceil(maxlen/50):
        break
    #         df2 = pd.DataFrame.from_dict({'MSA': seriesId, 'year': year, 'period': period, 'value': value})
    #         pd.concat([output_df, df2], sort = False)
    # output = open('output.txt','w')
    # output.write (output_df.get_string())
    # output.close()
```

```python
output_df = pd.DataFrame({'MSA': msa_vector, 'Year': year_vector, 'Period': period_vector, 'Value': value_vector})
```
